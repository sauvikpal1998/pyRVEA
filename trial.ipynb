```python
import numpy as np
import pygmo
import matplotlib.pyplot as plt
from pyrvea.Selection.constraint_select import constraint_tour_select
```

```python
# Booth function

def SimpleSqSum(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 + y**2

def SimpleSqDiff(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 - y**2

def Booth(arr):
    x = arr[0]
    y = arr[1]
    
    return (x+2*y-7)**2 + (2*x+y-5)**2

# Matayas
def Matayas(arr):
    x = arr[0]
    y = arr[1]
    
    return .26*(x**2+y**2)-0.48*(x*y)

# elif self.name == "Three-hump camel":
def ThreeHump(arr):
    x = arr[0]
    y = arr[1]
    return 2 * x ** 2 - 1.05 * x ** 4 + (x ** 6) / 6 + x * y + y ** 2


# elif self.name == "LeviN13":
def LeviN13(arr):
    x = arr[0]
    y = arr[1]

    
    temp = np.sin(3 * np.pi * x) ** 2 + (x - 1) ** 2 * (1 + np.sin(3 * np.pi * y) ** 2) + (y - 1) ** 2 * (1 + np.sin(2 * np.pi * y) ** 2)
        
    return temp


def const1(arr):
    x = arr[0]
    y = arr[1]
    val = x+y
    if(val>3 and val<40):
        return True
    else:
        return False
    
def const2(arr):
    x = arr[0]
    y = arr[1]
    val = x-y
#     print(val)
    if(abs(val)>3 and abs(val)<40):
        return True
    else:
        return False
    
def const3(arr):
    if(arr[1]<2):
        return True
    else:
        return False

def const4(arr):
    if(arr[1]>-2):
        return True
    else:
        return False
    
def const5(arr):
    if(arr[0] + arr[1] > -1):
        return True
    else:
        return False

# def const6(arr):
#     if(arr[0] + arr[1] > -3):
#         return True
#     else:
#         return False
```

```python
address = "pics/exp5/"
pop_size = 500

"""NAMING"""

obj1 = Matayas
obj2 = SimpleSqDiff

"""SEED"""

np.random.seed(5)

"""DATA GEN"""

values = np.random.uniform(-5, 5, (pop_size, 2))

vals1 = [obj1(v) for v in values]
vals2 = [obj2(v) for v in values]

x = values[:,0]
y = values[:,1]

values = values.tolist()

"""DATA GEN GRAPHS"""
print("X vs Y")
plt.title("X vs Y")
plt.scatter(x, y)
plt.savefig(f'{address}X vs Y.png')
plt.show()

print("f1 vs f2")
plt.title("f1 vs f2")
plt.scatter(vals1, vals2)
plt.savefig(f'{address}f1 vs f2.png')
plt.show()

"""OBJ INS"""

objectives = [obj1, obj2]
constraints = [const4, const3]
flags = ["max", "min"]
clf = constraint_tour_select(constraints, objectives, flags)


"""LOOPS"""

# pop = values

for i in range(10):
    print(f"------{i}-------")
    
    pop = clf.get_new_pop(values)
    vals1 = [obj1(v) for v in pop]
    vals2 = [obj2(v) for v in pop]
    
    pop_n = np.array(pop)
    x = pop_n[:, 0]
    y = pop_n[:, 1]
    
    print("x vs y")
    plt.title(f"x vs y, iter = {i+1}")
    plt.scatter(x, y)
    plt.savefig(f'{address}X vs Y_{i+1}.png')
    plt.show()
    
    print("f1 vs f2")
    plt.title(f"f1 vs f2, iter = {i+1}")
    plt.scatter(vals1, vals2)
    plt.savefig(f'{address}f1 vs f2_{i+1}.png')
    plt.show()
    values = pop
```
