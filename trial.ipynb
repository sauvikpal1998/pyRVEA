```python
# Booth function
def obj1(arr):
    x = arr[0]
    y = arr[1]
    
    return (x+2*y-7)**2 + (2*x+y-5)**2

# Matayas
def obj2(arr):
    x = arr[0]
    y = arr[1]
    
    return .26*(x**2+y**2)-0.48*(x*y)

def obj3(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 + y**2

def obj4(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 - y**2


def const1(arr):
    x = arr[0]
    y = arr[1]
    val = x+y
    if(val>3 and val<40):
        return True
    else:
        return False
    
def const2(arr):
    x = arr[0]
    y = arr[1]
    val = x-y
#     print(val)
    if(abs(val)>3 and abs(val)<40):
        return True
    else:
        return False
```
