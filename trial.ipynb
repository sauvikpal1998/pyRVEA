```python
import numpy as np
import pygmo
import matplotlib.pyplot as plt
```

```python
def obj1(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 + y**2

def obj2(arr):
    x = arr[0]
    y = arr[1]
    
    return x**2 - y**2

def const1(arr):
    x = arr[0]
    y = arr[1]
    val = x+y
    if(val>3):
        return True
    else:
        return False
    
def const2(arr):
    x = arr[0]
    y = arr[1]
    val = x-y
#     print(val)
    if(abs(val)>3):
        return True
    else:
        return False
```

```python
values = np.random.uniform(-5, 5, (200, 2))
values = values.tolist()
# vals1 = np.array([obj1(v) for v in values])
# vals2 = np.array([obj2(v) for v in values])
vals1 = [obj1(v) for v in values]
vals2 = [obj2(v) for v in values]
```

```python
from pyrvea.Selection.tournament_select import constraint_tour_select
objectives = [obj1, obj2]
constraints = [const1, const2]
```

```python
pop = constraint_tour_select(constraints, values, objectives)
```

```python
vals1 = [obj1(v) for v in pop]
vals2 = [obj2(v) for v in pop]
plt.scatter(vals1, vals2)
```
